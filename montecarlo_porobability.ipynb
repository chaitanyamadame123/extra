```python
import numpy.random as round
import numpy as np
import matplotlib.pyplot as plt
```

```python
def mc_normal(mean,std_dev,samples):
    results = []
    for _ in range(samples):
        results.append(np.random.normal(mean,std_dev))
    return np.array(results)    
```

```python
# configuration
s = 100000 # number of samples
upper_limit =34 #upper limit from specification
#components
components_1 = mc_normal(5,1,s)
components_2 = mc_normal(10,1,s)
components_3 = mc_normal(15,1,s)

#relationships
total = components_1+components_2+components_3
#success conditions
probability = np.sum(total>upper_limit)/len(total)*100
print("Probability of exceeding the time limit: ", np.round(probability,3),"%")

count,bins,ignored = plt.hist(total,30,density=True,color='black')
plt.vlines(34,count.min(),count.max(),color='r',label="34 Minute")
plt.legend();plt.show()
```
