```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x):
    return -(x**2)+(4*x)+5
```

```python
def f_prime(x):
    return -(2*x + 4)
```

```python
def f_double_prime(x):
    return np.ones(400)*(-2)
```

```python
x = np.linspace(-2, 6, 400)
```

```python
y = f(x)
y_prime = f_prime(x)
y_double_prime = f_double_prime(x)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(x, y)
plt.plot(x, y_prime)
plt.plot(x, y_double_prime)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of f(x), f\'(x), and f\'\'(x)')
plt.legend()
plt.grid(True)
plt.show()
```
